```python
from nltk.corpus import wordnet
synonyms = []
antonyms = []

for syn in wordnet.synsets("Military"):
    for l in syn.lemmas():
            synonyms.append(l.name())
            if l.antonyms():
                antonyms.append(l.antonyms()[0].name())
```

```python
investor_text=[]
investor_text.append('My name is John. I am 34. I work information technology sector. I would like to invest $30000 in my RRSP this year. I do not like having military and tobacco in my portfolio. I care a lot about sustainable development.')
investor_text.append('My name is Greg. I am 44. I work information technology sector. I would like to invest $30000 in my RRSP this year. I do not like gambling. I care a lot about sustainable development.')
investor_text.append('My name is Craig. I am 51. I work information technology sector. I would like to invest $30000 this year. I do not like stocks with fossil fuels. I care a lot about sustainable development.')
investor_text.append('My name is Jen. I am 28. I work information technology sector. I would like to invest $30000 this year. I do not like stocks with fossil fuels and tobacco.')
investor_text.append('My name is Jenny. I am 39. I work information technology sector. I would like to invest $30000 this year. I do not like stocks with fossil fuels and military assets.')
investor_text.append('My name is Tom. I am 29. I work information technology sector. I would like to invest $30000 in my RRSP this year. I do not like stocks with tobacco and I dont like gambling')
investor_text.append('My name is Jordan. I am 35. I work information technology sector. I would like to invest $30000 in my RRSP this year. I do not like stocks with fossil fuels and tobacco.')


```

```python
#Programmatically adding sentences with different sectors
sectors = ["information technology","health Care","government","consumer staples","real estate"]
for s in sectors:
    for sentence in investor_text:
            if sentence.find(s) != -1:
                for j in sectors:
                    if j!=s:
                        new_sentence = sentence.replace(s,j)
                        investor_text.append(new_sentence)
                        
```

```python
train_data = [
    ("Uber blew through $1 million a week", [(0, 4, 'ORG')]),
    ("Android Pay expands to Canada", [(0, 11, 'PRODUCT'), (23, 30, 'GPE')]),
    ("Spotify steps up Asia expansion", [(0, 8, "ORG"), (17, 21, "LOC")]),
    ("Google Maps launches location sharing", [(0, 11, "PRODUCT")]),
    ("Google rebrands its business apps", [(0, 6, "ORG")]),
    ("look what i found on google! ", [(21, 27, "PRODUCT")])]
```

```python
data=[]
curset={}
for sentence in split_text:
    for line in sentence:
        curset[line] = []
for flag in flags :
    synonyms = []
    for syn in wordnet.synsets(flag):
        for l in syn.lemmas():
            synonyms.append(l.name())

    synonyms = list(set(synonyms))
    cur_list=[]
    for x in curset:
        cur_list.append(x)

    for cur_line in cur_list:
        if cur_line.find(flag) != -1:
            for j in synonyms:
                new_line = cur_line.replace(flag,j)
                if new_line not in curset:
                    curset[new_line]=[]
                    for t in curset[cur_line]:
                        curset[new_line].append(t)
                curset[new_line].append((j,flag))

for data_line in curset:
    if len(curset[data_line])!=0 :
        flag_set=[]
        for cur_flag in curset[data_line]:
            flag_set.append((data_line.find(cur_flag[0]),data_line.find(cur_flag[0])+len(cur_flag[0]),cur_flag[1]))
        data.append((data_line,flag_set))
print(data)
```
